```python
def solve(n):
    num_door = n
    doors = [1 for num in range(num_door)]

    def flip(n):
        if doors[n - 1] ==1:
            doors[n - 1] = 0
        else:
            doors[n - 1] = 1
    
    for student in range(num_door):
        for door in range(num_door):
            if (student + 1) * (door + 1) <= num_door:
                flip((student + 1) * (door + 1))
        print(doors) 
```
